```python
def scroll(duartion=70): # allows you to scroll the Facebook. page for specified duration (seconds)
    
    global driver
    
    for i in range(duartion):
        
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        
        time.sleep(random.randint(3,4))
        
    driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
    
    driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
    
    time.sleep(random.randint(7,9))
```

```python
def like_last_five_posts(): # likes the last 5 posts of profile
    
    global driver
    
    scroll(10)
    
    likes=driver.find_elements_by_xpath('//span/div/div/span/a[contains(text(),"Like")]') # correct one
    
    for i in range(min(5,len(likes))): # number of posts to be liked: here 5
        
        try:
            
            likes[i].click()
            
            time.sleep(random.randint(7,9))
            
        except:
            
            pass
    
```

```python
 def send_messages(msg='Hello! How are you?'): # sends message to profiles default message: Hello! How are you?
        
        global driver
        
        scroll()
        
        time.sleep(10)
        
        try:
            
            messages=driver.find_elements_by_partial_link_text('Message')
            
            messages[1].click()
            
            time.sleep(3)
            
            txt=driver.find_elements_by_xpath('//div[@role="combobox"]')
            
            txt[0].click()
            
            txt[0].send_keys(msg)
            
            time.sleep(5)
            
            txt[0].send_keys(Keys.ENTER)
        
        except:
            
            pass
            
```

```python
def put_comment_on_post(range_start,range_end): # puts comments to posts of profiles
    
    global driver
    
    global range_start
    
    global range_end
    
    scroll(10)
    
    com=random.randint(range_start,range_end)
    
    comment_to_post=your_comments[com]
    
    try:
        
        comment=driver.find_elements_by_xpath('//a[@title="Leave a comment"]')
        
        if len(comment)>5:
            
            comment=comment[:5]

        for com in comment:

            try:
                
                com.click()
                
                time.sleep(random.randint(3,5))

            except:pass
            
        driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
        
        driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
        
        time.sleep(random.randint(3,5))

        text=driver.find_elements_by_xpath('//div[@role="textbox"]')
        
        text=text[1:]
        
        if len(text)>5:
            
            text=text[:5]
            
        for i in range(len(text)):
            
            try:

                text[i].send_keys(comment_to_post)

                text[i].send_keys(Keys.ENTER)
                
                time.sleep(random.randint(100,120))
                
            except:
                
                pass
            
    except:
        
        pass
       
```
